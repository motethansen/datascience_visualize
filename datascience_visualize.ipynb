# Exploring visualization with public data

Source: https://www.singstat.gov.sg/find-data/search-by-theme/industry/transport/latest-data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
mrt_data =[(2022,2745000),
           (2021,2100000),
           (2020,2023000),
           (2019,3384000),
           (2018,3302000),
           (2017,3122000),
           (2016,3095000),
           (2015,2871000),
           (2014,2762000),
           (2013,2623000),
           (2012,2525000),
           (2011,2295000),
           (2010,2069000),
           (2009,1782000),
           (2008,1698000),
           (2007,1527000),
           (2006,1408000),
           (2005,1321000)] 
```

```python
df = pd.DataFrame(mrt_data, columns =['Year', 'passengers'])
df.head()
```

```python
df["Year"] = df["Year"].values.astype(str)
df
```

```python
# passengers in thousand
df["passengers"] = df["passengers"]/1000
df.head(3)
```

```python
#sort the list
df.sort_values(by=['Year'], inplace=True)
df
```

```python
plt.figure(figsize = (9,6))
plt.bar(x=df["Year"],
        height = df["passengers"],
       color = "midnightblue")
plt.xticks(rotation=45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.ylabel("Number of passenger in thousand", fontsize = 13)
#plt.xlabel("Year")
plt.title("Number of passengers traveling by MRT per year", fontsize = 16, fontweight ="bold")
plt.savefig("passengers_mrt.png")
plt.show()
```
